## The steps for the project are:

- Import necesary packages (e.g. matplotlib and numpy etc.)
- Generate some random sample data
- Plot the random data using matplot lib

```python
# Import packages
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Generate random data
x = np.arange(0.0, 12.0, 0.1)
y = -3.0 * x + 50
noise = np.random.normal(0.0, 1.0,len(x))

# Plot the data, then the line
plt.plot(y + noise, x, 'r.', label='Data')
plt.plot(y, x, 'b-', label='Linear')
plt.title("Cooker setting vs. Time to Boil water")
plt.ylabel("Cooker setting")
plt.xlabel("Time to Boil Water")
plt.legend()
plt.show()
```
