# K Nearest Neighbors and the Iris Data Set

This week in the lectures for both FODA and PFDA we covered an introduction to machine learing with the SCIKIT package in Python. We used the KNN alogorithm on the iris data set as an example

## Load the Data Set

```python
# Import Pandas and Seaborn
import pandas as pd
import seaborn as sns
import sklearn.neighbors as nei
import sklearn.model_selection as mod
# Load the data set as a datafram from Ians github
df = pd.read_csv("https://github.com/ianmcloughlin/datasets/raw/master/iris.csv")

# Display the data set
df
```

## Visualise the data set with Seaborn

```python
#Use a pairplot
sns.pairplot(df,hue="class")
```

## Inputs and Outputs

```python
inputs = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
outputs = df['class']
```

## Classifier

Create at classifer object that can be trained with our training data set

```python
knn = nei.KNeighborsClassifier(n_neighbors=5)
```

## Fit

This step trains the classifier with our data set. We need to show the classifier what the inputs and outputs are.

```python
knn.fit(inputs,outputs)
```

## Predict

Pick a enter some random values and get a prediction from the classifier

```python
knn.predict([[4.1, 1.2, 2.5, 1.3]])
```

Note we can add multiple flowers

```python
knn.predict([[4.1,1.2,2.5,1.3],[1.1, 3.4, 5.6, 1.2]])
```

Note - it can take a an array input if we feed it the "input" data that it was trained on we can see that it fudges a little with the virginicas and versicolors!

```python
knn.predict(inputs)
```

## Evaluate

This step is used to evaluate the data set by feeding the classifier the training set, and then comparing the output from the classifier to the actual classifications (our "outputs" array)

```python
# Evalaute by create a pandas df of true/false items to see how they compare
evaluate = knn.predict(inputs) == outputs
# Sum up the numer of true items
evaluate.sum()
# Get the percentage
perc_acc = evaluate.sum()*100/evaluate.count()
print("The classifier is " + str(round(perc_acc,2)) + " accurate")
```

Another way to evaluate the classifier is to split the data set into training and test data randomly (theres a sklearn function to do it!). Then train the classifier with some of the data and then test it with the rest. We can then evaluate how the classifer worked on the test data set.

```python
# Create the test and trial inputs and outputs
input_trial, input_test, output_trial, output_test = mod.train_test_split(inputs, outputs, test_size=0.33)

# Train the classifier with the training set
knn = nei.KNeighborsClassifier(n_neighbors=5)
knn.fit(input_trial,output_trial)

# Test it with the test data
test = knn.predict(input_test) == output_test
test.sum()

print("The percentage of correctly classified test items is " + str(round((test.sum()*100/test.count()),2)) + "%")
```
